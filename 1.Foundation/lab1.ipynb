## 1ST AI basic Agent Code

```python
prompt = "Give me such an Domain or  idea for bussiness area that might be so new for an Agentic AI startup that is never ever exploited or people never worked on that part "
response = groq.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": prompt }]
)



business_idea = response.choices[0].message.content

response=groq.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content" : business_idea + "Pain point in that industry - something challenging that might be ripe fpr Agentic Solution."}]
)
answer=response.choices[0].message.content

print(answer)
```
